```python
# creating a DataFrame for each month of 2018
df_Jan2017 = pd.read_csv("JC-201701-citibike-tripdata.csv")
df_Feb2017 = pd.read_csv("JC-201702-citibike-tripdata.csv")
df_Mar2017 = pd.read_csv("JC-201703-citibike-tripdata.csv")
df_Apr2017 = pd.read_csv("JC-201704-citibike-tripdata.csv")
df_May2017 = pd.read_csv("JC-201705-citibike-tripdata.csv")
df_Jun2017 = pd.read_csv("JC-201706-citibike-tripdata.csv")
df_Jul2017 = pd.read_csv("JC-201707-citibike-tripdata.csv")
df_Aug2017 = pd.read_csv("JC-201708-citibike-tripdata.csv")
df_Sep2017 = pd.read_csv("JC-201709-citibike-tripdata.csv")
df_Oct2017 = pd.read_csv("JC-201710-citibike-tripdata.csv")
df_Nov2017 = pd.read_csv("JC-201711-citibike-tripdata.csv")
df_Dec2017 = pd.read_csv("JC-201712-citibike-tripdata.csv")
```
